```python
import speech_recognition as sr #This library is used to recognize speech and convert it into text. 
import webbrowser #This is a built-in Python module that provides a simple interface to display web pages in a web browser.
import pyttsx3 #This is a text-to-speech (TTS) library that converts text into spoken words using your system’s built-in speech synthesis engine.
import pywhatkit as kit #Python library that provides several useful functionalities like performing Google searches, sending WhatsApp messages, playing YouTube videos, and more. It simplifies automation tasks that require interaction with different websites and apps.
```

```python
engine=pyttsx3.init() #This initializes the pyttsx3 engine, which will convert text into speech
def speak(text): #A function that takes a string (text) as input and makes the program speak that text.
    engine.say(text) #Converts the text into speech.
    engine.runAndWait() #Runs the speech engine and waits for the speech to finish before proceeding.
def processcommand(c): #A function that processes a command (c) received via speech and executes actions based on the recognized command
    if "open google" in c.lower():
        webbrowser.open("https://google.com")
    elif "open youtube" in c.lower():
        webbrowser.open("https://youtube.com")
    elif "open facebook" in c.lower():
        webbrowser.open("https://facebook.com")
    elif "open linkedin" in c.lower():
        webbrowser.open("https://linkedin.com")
    elif "play" in c.lower():
        kit.playonyt(c)
    elif "search" in c.lower():
        search_query = c.replace("search", "") #search is replaced with empty string to avoid engine for enterpretting search as a word in search bar.
        kit.search(search_query)
    
if __name__=="__main__":
    speak("initialising jarvis....") #The program speaks out "initializing jarvis" to indicate it is starting.
    while True:
        r=sr.Recognizer() #creating an instance of the Recognizer class from the speech_recognition
        try:
            with sr.Microphone() as source: #This opens the microphone to start listening
                print("listening....")
                audio = r.listen(source,timeout=2,phrase_time_limit=3 ) #This listens for audio input with a timeout of 2 seconds and a phrase time limit of 3 seconds.
                print("recognizing....")
            word=r.recognize_google(audio) #This sends the recorded audio to Google's speech recognition API to convert it to text.
            if(word.lower()=="jarvis"): #If the recognized word is "jarvis" (case-insensitive), the program speaks "yes sir" and continues to listen for the next command.
                speak("yes sir")
                with sr.Microphone() as source: #Opens the microphone again to listen for a command after hearing "Jarvis".
                    print("jarvis active....")
                    audio = r.listen(source,timeout=2,phrase_time_limit=3) #The program listens for a new command from the user.
                command=r.recognize_google(audio) #Converts the new audio input into text.

                processcommand(command) #function call

                if command.lower() == "stop":  # Check if the command is "stop"
                    print("jarvis stopped")
                    speak("Stopping Jarvis. Goodbye!")  #Say goodbye
                    break 
                
                
        except sr.WaitTimeoutError: # This error occurs if no speech is detected within the timeout period.
            print("Listening timed out while waiting for phrase to start")
        except sr.UnknownValueError: #If the speech could not be understood by the speech recognition API, this error is raised.
            print("Google Web Speech API could not understand audio")
        except sr.RequestError as e: #If there’s an issue with the connection to the Google Speech API, this error will be raised.
            print(f"Could not request results from Google Web Speech API; {e}")
        except Exception as e:
            print(f"Error: {e}") #This catches any other unexpected errors and prints the error message.
```
